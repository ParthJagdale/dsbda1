```python
y = df.iloc[:, 4].values
y
```

# Label Encoding

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y

# Splitting data into 70% training and 30% testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

# Initialize Naive Bayes Model and training it

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
```

# Prediction

```python
prediction = model.predict(x_test)
prediction
```

```python
y_test
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, prediction)
cm
```

# Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, prediction)*100
accuracy
```

# Precision

```python
from sklearn.metrics import precision_score
precision = precision_score(y_test, prediction, average = 'micro')*100
precision
```

# Recall

```python
from sklearn.metrics import recall_score
recall = recall_score(y_test, prediction, average = 'micro')*100
recall
```
