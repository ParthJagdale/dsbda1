```python
df = pd.read_csv('student_performance1.csv')
```

```python
df['math score'] = df['math score'].interpolate()
df
```

```python
df['Reading Score'] = df['Reading Score'].interpolate()
```

```python
df.isnull().sum()
```

```python
df['Writing Score'].fillna(method = 'ffill', inplace = True)
```

```python
df.isnull().sum()
```

```python
df['Writing Score'].fillna(method = 'bfill', inplace = True)
```

```python
df.isnull().sum()
```

```python
df['Placement Score'] = df['Placement Score'].interpolate()
```

```python
df.isnull().sum()
```

```python
df['Club join Date'] = df['Club join Date'].replace(to_replace=np.nan, value = 2)
```

```python
df.isnull().sum()
```

```python
df['placement offer count'] = df['placement offer count'].replace(to_replace=np.nan, value = 1)
df
```
